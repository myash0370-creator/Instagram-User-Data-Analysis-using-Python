Collection of Data from instagram users 

```python

with open("initialdata.txt" ,encoding = 'utf8') as f:
    data = f.read()
# data
    
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
# chunks
```

```python
chunks[4]
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split('\n')

    username = sep_chunk[0]
    name = sep_chunk[1]

    no_of_posts = ""
    no_of_followers = ""
    no_of_following = ""

    # correct mapping
    for item in sep_chunk:
        if "post" in item:
            no_of_posts = item
        elif "followers" in item:
            no_of_followers = item
        elif "following" in item:
            no_of_following = item

    if len(sep_chunk) > 5:
        type_of_page = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
        type_of_page = "Unknown"
        bio = ""

    return {
        "username": username,
        "no_of_posts": no_of_posts,
        "no_of_followers": no_of_followers,
        "no_of_following": no_of_following,
        "name": name,
        "type_of_page": type_of_page,
        "bio": bio
    }

# print(parse_chunk(chunks[4]))
```

```python
all_chunks = [] 
for chunk in chunks:
    # print(f"The chunks is: {chunks}")
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
# print(all_chunks)
```

```python
import json
s = json.dumps(all_chunks, indent=4)
# print(s)
```

#Finding People with Max Posts and Followers

```python
def extract_number(text):
    num = text.split()[0].replace(",", "").lower()

    if "k" in num:
        return int(float(num.replace("k", "")) * 1000)
    elif "m" in num:
        return int(float(num.replace("m", "")) * 1_000_000)
    else:
        return int(float(num))


max_posts = 0
max_posts_user = ""

max_followers = 0
max_followers_user = ""


for ch in chunks:
    data = parse_chunk(ch)

    posts = extract_number(data["no_of_posts"]) if data["no_of_posts"] else 0
    followers = extract_number(data["no_of_followers"]) if data["no_of_followers"] else 0

    if posts > max_posts:
        max_posts = posts
        max_posts_user = data["username"]

    if followers > max_followers:
        max_followers = followers
        max_followers_user = data["username"]


print("Max Posts:", max_posts, "by", max_posts_user)
print("Max Followers:", max_followers, "by", max_followers_user)
```
